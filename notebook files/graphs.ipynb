```python
########################This code write the correlation matrix into a csv file.###########################

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("E:\\final\\all.csv", header = 0)
df = df.dropna()

corr = df.corr()
plt.figure(figsize = (20,20))

ax = sns.heatmap(corr, vmin=-1, vmax = 1, center=0, cmap=sns.diverging_palette(50,400,n=200), square=True)
plt.savefig('coor_matrix.png')

#corr.style.background_gradient(cmap='coolwarm')

#corr.to_csv("E:\\final\\correlation_matrix.csv",index=True)
# 'RdBu_r' & 'BrBG' are other good diverging colormaps
```
